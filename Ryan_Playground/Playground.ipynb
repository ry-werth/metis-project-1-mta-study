```python
df["DATE"] = pd.to_datetime(df["DATE"])
```

```python
df["DAY"] = df["DATE"].dt.dayofweek
```

```python
df.head()
```

```python
day_dict = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}
```

```python
df["DAY_STR"] = df["DAY"].map(day_dict)
df.head()
```

```python
df[:1000]
```

```python
df['ENTRIES_DIFF']=df.groupby(['STATION', 'SCP','UNIT'])['ENTRIES'].diff().fillna(0)
df.head()
```

```python
df.columns
```

```python
df = df.rename(columns={'EXITS                                                               ': 'EXITS'})
```

```python
df['EXIT_DIFF']=df.groupby(['STATION', 'SCP', 'UNIT'])['EXITS'].diff().fillna(0)
df.head()
```
